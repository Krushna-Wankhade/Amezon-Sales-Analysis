```python
#check b2b data
check_B2B=df['B2B'].value_counts()

#create pie chart
plt.pie(check_B2B, labels=check_B2B, autopct='%1.1f%%')
plt.show()
```

```python
#scatter plot
x_data= df['Category']
y_data= df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available size')
plt.show()
```

```python
#top 10 states
top10_state= df['ship-state'].value_counts().head(10)  #for top 10 states
plt.figure(figsize=(10,6))
plt.xticks(rotation=30)
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)],x="ship-state")
plt.xlabel('State')
plt.ylabel('Order_count')
plt.title('Distribution of State')
plt.show()
```
